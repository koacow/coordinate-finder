```python
import pandas as pd
import requests
import googlemaps as gmaps

# Change 'Pathology_locations.csv' to path of your file
input_file_path = 'Pathology_locations.csv'
data = pd.read_csv(input_file_path)[['Type of location', 'Facility', 'Latitude', 'Longitude']] 
```

```python
def get_lat_long_from_place_name(placeName):
    query = str(placeName)
    locations = []
    try:
        result = requests.post('https://places.googleapis.com/v1/places:searchText',
                               params={
                                   'fields': 'places.location',
                                   'textQuery': query,
                                   },
                               headers={'Content-Type': 'application/json', 'X-Goog-Api-Key': '<INSERT_API_KEY>'})
        json = result.json()
        return json['places'][0]['location']
    except Exception as e:
        print(f'Error: {e}')
        return {'latitude': 'None', 'longitude': 'None'}
    
dataLength = len(data)

for i in range(dataLength):
    print(f'Processing {i+1} of {dataLength}')
    location = get_lat_long_from_place_name(data['Facility'][i])
    if (location['longitude'] == 'None' or location['latitude'] > 0 or location['longitude'] < 100):
        data.loc[i, 'Latitude'] = 'None'
        data.loc[i, 'Longitude'] = 'None'
    else:
        data.loc[i, 'Latitude'] = location['latitude']
        data.loc[i,'Longitude']= location['longitude']
```

```python
# Change 'Pathology_locations_with_lat_long.xlsx' to desired output file name
export_file_name = 'Pathology_locations_with_lat_long.xlsx'
data.to_excel(export_file_name, index=False)
```
